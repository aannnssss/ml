7. Рассмотрим случайный вектор с формой 100х2, представляющей координаты, и найдем расстояния от точки к точке

````python
coordinates = np.random.randint(-50, 51, (100, 2)) 
distances = np.zeros((100, 100))

def euclidean_distance(p1, p2):
    return np.sqrt(np.sum((p1 - p2) ** 2))

for i in range(100):
    for j in range(100):
        distances[i, j] = euclidean_distance(coordinates[i], coordinates[j])

distances = np.round(distances, 2)

print(coordinates)
print()
print(distances)
````

1. Посчитайте количество записей в датасете с отсутствующей информацией о жилой площади жилого помещения

````python
missing_life_sq = df['life_sq'].isna().sum()

print(missing_life_sq)
````

2. Предложите способ заменить отсутствующие значения из пункта 1 и реализуйте его. Способ нужно аргументировать(сказать почему это достаточно хорошее приближение)

````python
df['life_sq'] = df['life_sq'].fillna(df.groupby('sub_area')['life_sq'].transform('mean'))

'''
Способ: замена отсутствующих значений на среднюю жилую площадь жилого помещения по району.
Это является достаточно хорошим приближением,
так как учитывает особенности района и поэтому будет лучше, 
чем просто медиана по столбцу 'life_sq'.
'''
````

9. Проанализируйте есть ли разница в цене между квартирами на 1 и 9 этажах

### 3. Визуализация

Используя библиотеки matplotlib/sns/plotly/pandas и набор данных представленный ранее визуализируйте

````python
import matplotlib.pyplot as plt
import seaborn as sns
````

1. Как менялась средняя цена квартиры в Нагатинском Затоне(Nagatinskiy Zaton) с течением времени

````python
df_nagatinskiy = df[df['sub_area'] == 'Nagatinskij Zaton']

mean_prices = df_nagatinskiy.groupby(pd.to_datetime(df_nagatinskiy['timestamp']).dt.strftime('%Y'))['price_doc'].mean()

plt.figure(figsize=(10, 6))
sns.lineplot(x=mean_prices.index, y=mean_prices.values)
plt.xlabel('Год')
plt.ylabel('Средняя цена квартиры')
plt.title('Изменение средней цены квартиры в Нагатинском Затоне с течением времени')
plt.show()
````

7. Постройте график изменения медианной цены на квартиру менее 50 кв. метров за период с 2011 до 2014 года*

\* Задание на доп. балл

````python

df['timestamp'] = pd.to_datetime(df['timestamp'])
df_filtered = df[(df['full_sq'] < 50) & (df['timestamp'].dt.year.between(2011, 2014))]
median_prices = df_filtered.groupby(df_filtered['timestamp'].dt.strftime('%Y'))['price_doc'].median()

plt.figure(figsize=(6, 6))
plt.plot(median_prices.index, median_prices.values, marker='o', linestyle='-')
plt.title('Изменение медианной цены на квартиры менее 50 кв.м. за 2011-2014 года')
plt.xlabel('Год')
plt.ylabel('Медианная цена')
plt.show()
````
